```python
f, ax = plt.subplots(figsize=(10,10))

meanwidth = 0.3

for x, nt in enumerate(df.columns):
    ax.scatter([x]*len(df), df[nt])
    ax.plot([x-meanwidth, x+meanwidth], [df[nt].median(), df[nt].median()], color="black", linewidth=2, alpha=0.5)

ax.set_xticks(range(len(df.columns)))
ax.set_xticklabels(df.columns, rotation=45)
```

```python
import numpy as np

plt.rcParams['font.size'] = 18

f, ax = plt.subplots(figsize=(2.3,2.8),
                     gridspec_kw={"left":0.02, "right":0.98, "top":0.98, "bottom":0.02})

hist_color = "#0097A3"
other_color = "grey"

colors = [other_color] * 7 + [hist_color]
meanwidth = 0.4

def jitter_y_vals(y, jitter_strength=0.2):
    return y + np.random.uniform(-jitter_strength, jitter_strength, size=len(y))

for (y, nt), color in zip(enumerate(df.columns), colors):
    ax.scatter(df[nt], jitter_y_vals([y]*len(df)), color="white", alpha=0.4, s=40, edgecolor=color, linewidth=2)
    ax.plot([df[nt].median(), df[nt].median()], [y-meanwidth, y+meanwidth], color="black", linewidth=1, zorder=0, alpha=0.7)
    

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.vlines([0,50,100,150,200], -0.5, 8, color="black", alpha=0.2, linestyle="--", zorder=0)

ax.set_yticks(range(len(df.columns)))
ax.set_ylim(-0.5, 7.5)
ax.set_yticklabels([])
ax.set_xticklabels([])

f.savefig(FIGSFOLDER / "in_vitro.pdf")
```
