```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/kvo019/OneDrive - UiT Office 365/Pipette 2/People/Linnea/Manuscripts/hislight paper/stats/behavior_data_ALL.csv")
```

```python
columns_to_test = ['restricted_licks', 'dark_licks', 'light_licks', 'sucralose_licks']
# Perform Shapiro-Wilk test for each column
for column in columns_to_test:
    stat, p_value = shapiro(data[column])
    print(f"Column: {column}")
    print(f"Shapiro-Wilk Test Statistic: {stat}")
    print(f"P-value: {p_value}")
    if p_value > 0.05:
        print("The data are normally distributed")
    else:
        print("The data are NOT normally distributed")
    print("-" * 50)
```
